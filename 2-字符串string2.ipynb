了解**单引号、双引号和三引号**之间的区别。

- 使用单引号创建字符串时，如果字符串中包含单引号字符，则必须在单引号字符前加上转义符“\”；

- 而使用双引号创建字符串时，如果字符串中包含双引号字符，则必须在双引号字符前加上转义符“\”。

因此，我们可以根据实际情况决定创建字符串时使用哪种引号，从而在编写代码时可以减少转义符的使用、增强程序的可读性。

````python
str1='It\'s a book.' #使用\'说明其是字符串中的一个单引号字符，不加\则会报错
str2="It\'s a book." #使用一对双引号创建字符串，此时字符串中的单引号不需要转义符
str3="He said:\"It is your book.\"" #使用\"说明其是字符串中的双引号字符
str4='He said:"It is your book.' #使用一对单引号创建字符串，省掉了转义符
print(str1) #输出str1
print(str2) #输出str2
print(str3) #输出str3
print(str4) #输出str4
````

单引号和双引号中的字符串如果分多行写，必须在每行结尾加上续行符“\”；如果希望一个字符串中包含多行信息，则需要使用换行符“\n”。

````python
# 问题解决：凯撒加密，每个字母替换为后面第k个。
import string
def kaisa(s, k,decode = False):
    lower = string.ascii_lowercase          #小写字母
    upper = string.ascii_uppercase          #大写字母
    before = string.ascii_letters
    after = lower[k:] + lower[:k] + upper[k:] + upper[:k]
    if not decode:
        table = ''.maketrans(before, after)     #创建映射表
    else:
        table = ''.maketrans(after, before) # decode_table
    return s.translate(table)

s = "Python is a greate programming language. I like it!"
encode_s = kaisa(s, 9)
encode_s
kaisa(encode_s, 9,decode = True)


````

检查并判断密码字符串的安全强度。

````python
import string

def check(pwd):
    #密码必须至少包含6个字符
    if not isinstance(pwd, str) or len(pwd)<6:
        return 'not suitable for password'
    #密码强度等级与包含字符种类的对应关系
    d = {1:'weak', 2:'below middle', 3:'above middle', 4:'strong'}
    #分别用来标记pwd是否含有数字、小写字母、大写字母和指定的标点符号
    r = [False] * 4
    for ch in pwd:
        #是否包含数字
        if not r[0] and ch in string.digits:
            r[0] = True
        #是否包含小写字母
        elif not r[1] and ch in string.ascii_lowercase:
            r[1] = True
        #是否包含大写字母
        elif not r[2] and ch in string.ascii_uppercase:
            r[2] = True
        #是否包含指定的标点符号
        elif not r[3] and ch in ',.!;?<>':
            r[3] = True
    #统计包含的字符种类，返回密码强度
    return d.get(r.count(True), 'error')

print(check('a2d9090,'))
print(check('123456'))
````

最简单的方式就是循环迭代。最高效的字符串匹配算法是[KMP算法](https://www.bilibili.com/video/BV1AY4y157yL?t=6.6)。

什么是KMP算法：

KMP是三位大牛：D.E.Knuth、J.H.Morris和V.R.Pratt同时发现的。其中第一位就是《计算机程序设计艺术》的作者！！

KMP算法要解决的问题就是在字符串（也叫主串）中的模式（pattern）定位问题。说简单点就是我们平时常说的关键字搜索。模式串就是关键字（接下来称它为P），如果它在一个主串（接下来称为T）中出现，就返回它的具体位置，否则返回-1（常用手段）


````python
def kmp_search(text, pattern):
    M = len(pattern)
    N = len(text)

    # 构建前缀函数
    lps = [0] * M
    compute_lps(pattern, M, lps)

    i = j = 0
    while i < N:
        if pattern[j] == text[i]:
            i += 1
            j += 1

        if j == M:
            print("Found pattern at index " + str(i - j))
            j = lps[j - 1]

        # 不匹配且j非0，移动j到lps[j-1]
        elif i < N and pattern[j] != text[i]:
            if j != 0:
                j = lps[j - 1]
            else:
                i += 1

def compute_lps(pattern, M, lps):
    length = 0  # length of the previous longest prefix suffix
    lps[0] = 0  # lps[0] is always 0
    i = 1

    while i < M:
        if pattern[i] == pattern[length]:
            length += 1
            lps[i] = length
            i += 1
        else:
            if length != 0:
                length = lps[length - 1]
            else:
                lps[i] = 0
                i += 1

# 测试代码
text = "ABABDABACDABABCABAB"
pattern = "ABABCABAB"
kmp_search(text, pattern)
````

## 字符串的查找
### 2-正则表示式查找


最直接的方式就是使用if语句来循环匹配和查找，不用正则表达式来查找文本模式。

比如：假设你希望在字符串中
查找电话号码。你知道模式：3 个数字，一个短横线，3
个数字，一个短横线，再是 4 个数字。例如：415-555-4242。



````python
# 循环迭代的字符串查找模式

def isPhoneNumber(text):
    if len(text) != 12:
        return False
    for i in range(3):  
        if not text[i].isdecimal():
            return False
    if text[3] != '-':
        return False
    for i in range(4, 7):  
        if not text[i].isdecimal():
            return False
    if text[7] != '-':
        return False
    for i in range(8, 12):  
        if not text[i].isdecimal():
            return False
    return True

print('415-555-4242 is a phone number:')
print(isPhoneNumber('415-555-4242'))  # 修正了原代码中的括号
print('Moshi moshi is a phone number:')
print(isPhoneNumber('Moshi moshi'))  # 修正了原代码中的括号
````

使用正则表达式会特别简单

````python
import re

phoneNumRegex = re.compile(r'(\d\d\d)-(\d\d\d-(\d\d\d\d))')
# phoneNumRegex = re.compile(r'\d\d\d-\d\d\d-\d\d\d\d)')
mo = phoneNumRegex.search('My number is 415-555-4242.')
print('Phone number found: ' + mo.group(3))
````
