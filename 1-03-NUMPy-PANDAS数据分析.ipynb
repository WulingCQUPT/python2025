```python
df = pd.DataFrame(np.random.randn(500, 3))
```

```python
df.describe()
```

```python
outliers = df.copy()
mean = df.mean()
std = df.std()
for col in df.columns:
    outliers[col] = (df[col] > mean[col] + std[col] * 3) | (df[col] < mean[col] - std[col] * 3)
```

```python
df[outliers.any(1)]
```

```python
df[outliers] = 3 * np.sign(df)
```

### 8.2 数据分组与聚合

#### 8.2.1 数据分组

```python
df = pd.DataFrame({'data': np.random.randint(0, 10, size = 5),
                  'key1': ['A', 'B', 'B', 'A', 'A'],
                   'key2': ['a'] * 2 + ['b'] * 3})
```

```python
df
```

```python
df.groupby('key1')
```

```python
df.groupby('key1').median()
```

```python
df.groupby(['key1', 'key2']).std()
```

```python
key3 = ['T', 'T', 'P', 'T', 'P']
df.groupby(key3).median()
```

```python
mapping = {0: 'T', 1: 'T', 2: 'P', 3: 'T', 4: 'P'}
df.groupby(mapping).median()
```

```python
index = [['A'] * 3 + ['B'] * 3 + ['C'] * 3, ['i', 'ii','iii'] * 3]
mulidx_ser = pd.DataFrame(np.random.randn(9, 2), index = index)
mulidx_ser.index.names = ['One', 'Two']
```

```python
mulidx_ser
```

```python
mulidx_ser.groupby(level = 'One').median()
```

```python
mulidx_ser.groupby(level = 1).median()
```

```python
for group_name, group_data in df.groupby('key1'):
    print(group_name)
    print(group_data)
```

```python
for (name1, name2), group_data in df.groupby(['key1', 'key2']):
    print((name1, name2))
    print(group_data)
```
